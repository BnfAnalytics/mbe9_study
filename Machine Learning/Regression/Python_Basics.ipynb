# 3. Variables and Data Types
In Python, variables are used to store data. They don't need explicit declaration. Let's look at some basic data types:

```python
# Integer
num1 = 5

# Float
num2 = 3.14

# String
name = "John"

# Boolean
is_true = True

# Print the values
print(num1)
print(num2)
print(name)
print(is_true)
```

# 4. Operators
Python supports various operators for performing operations on variables:

```python
# Arithmetic operators
result = 5 + 3
print("Addition:", result)

result1 = 5 * 3
print("Multiplication:", result1)

# Comparison operators
is_equal = (5 == 3)
print("Is equal:", is_equal)

# Logical operators
logical_result = (True and False)
print("Logical AND:", logical_result)
```

```python
5+3
```

# 5. Control Flow
Control flow statements allow you to control the order of execution in your program. Here are some examples:

```python
# If statement
num = 10
if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")

# While loop
count = 0
while count < 5:
    print(count)
    count += 1

# For loop
for i in range(5):
    print(i)
```

# 6. Functions
Functions are reusable blocks of code. Here's a simple example:

```python
print("Hello Python")
```

```python
# functions.py

def greet(name):
    print("Hello, " + name + "!")

# Call the function
greet("Alice")
```

# 7. Lists
Lists are a versatile data type in Python that can hold multiple values:

```python
my_list = [1, 2, 3, "four", 5.0]
```

```python
my_list
```

```python
# Accessing elements
print(my_list[0])

# Slicing
print(my_list[1:4])

# Modifying elements
my_list[3] = "Ten"

# Adding elements
my_list.append(6)

# Removing elements
my_list.remove("Ten")

print(my_list)
```

# 8. Loops
Loops are used to iterate over a sequence. Here's an example using a for loop:

```python
# For loop
for i in range(5):
    print(i)
print("*****************")
# While loop
count = 0
while count < 5:
    print(count)
    count += 1
```

# 9. Conditionals
Conditionals are used to make decisions in your code:

```python
num = 10

if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")
```

# 10. Dictionaries
Dictionaries allow you to store key-value pairs:

```python
# Creating a dictionary
person = {
    "name": "John",
    "age": 25,
    "city": "New York"
}

# Accessing values
print(person["name"])
print(person.get("age"))

# Modifying values
person["age"] = 26

# Adding new key-value pairs
person["gender"] = "Male"

# Removing a key-value pair
del person["city"]

print(person)
```

# 11. Strings
Strings are sequences of characters in Python. Here's how to work with strings:

```python
my_string = "Hello, World!"

# String length
print(len(my_string))

# Concatenation
greeting = "Hello"
name = "Alice"
full_greeting = greeting + ", " + name + "!"
print(full_greeting)

# String methods
print(my_string.upper())
print(my_string.lower())
print(my_string.replace("Hello", "Hi"))
```

# 12. File Handling
Python provides easy ways to work with files:

```python
# Writing to a file
with open("example.txt", "w") as file:
    file.write("Hello, Python!")

# Reading from a file
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```
