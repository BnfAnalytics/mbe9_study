```python
data['Reviews']= data['Reviews'].astype(int)
```

```python
data['Installs']= data['Installs'].astype(int)
```

```python
data['Installs']=data['Installs'].str.replace('+','')
data['Installs']=data['Installs'].str.replace(',','')
```

```python
data['Installs']= data['Installs'].astype(int)
```

```python
data.dtypes
```

```python
data['Price']= data['Price'].astype(float)
```

```python
data['Price']=data['Price'].str.replace('$','')
data['Price']= data['Price'].astype(float)
```

```python
data.dtypes
```

```python
data['Size']= data['Size'].astype(float)
```

```python
data['Size']
```

```python
def change_size(size):
  if 'M' in size:
    x=size[:-1] #19M
    x=float(x)*1000
    return x
  elif 'k' in size:
    x=size[:-1] #19k
    x=float(x)
    return x
  else:
    return None
```

```python
data['Size']= data['Size'].apply(change_size)
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data['Size'].fillna(method='ffill',inplace=True)
```

```python
inp1=data.copy()
```

```python
inp1.head()
```

```python
inp1.drop(['App','Last Updated','Current Ver','Android Ver'],axis=1,inplace=True)
```

```python
inp1.head()
```

```python
inp1.shape
```

# Convert Categorical data into numerical data

```python
inp1=pd.get_dummies(inp1)
```

```python
inp1.shape
```

```python
inp1.head()
```

# Extract Independent and dependent variables

```python
y=inp1.pop('Rating')
x=inp1
```

```python
# Separate the dataframe into X_train, X_test, y_train and y_test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

# Apply Linear Regression

```python
from sklearn.linear_model import LinearRegression
linear_reg= LinearRegression()
```

```python
linear_reg.fit(X_train,y_train) # Train your data with Linear Regression model
```

```python
y_pred=linear_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print('MSE: ',mean_squared_error(y_test,y_pred))
print('R2 Score: ',r2_score(y_test,y_pred))
```
