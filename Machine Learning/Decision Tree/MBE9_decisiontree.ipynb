```python
print(dt.EmployeeCount.unique())
print(dt.Over18.unique())
print(dt.StandardHours.unique())
print(dt.EmployeeNumber.unique())
```

```python
dt=dt.drop(['EmployeeCount','Over18','StandardHours','EmployeeNumber'],axis=1)
```

```python
dt.shape
```

```python
dt.describe()
```

```python
print(dt.BusinessTravel.value_counts())
print('****************************************')
print(dt.Department.value_counts())
print('****************************************')
print(dt.EducationField.value_counts())
print('****************************************')
print(dt.Gender.value_counts())
print('****************************************')
print(dt.JobRole.value_counts())
print('****************************************')
print(dt.MaritalStatus.value_counts())
```

```python
# Encoding categorical variables into numerical variables
dt_dummies=pd.get_dummies(dt[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime']],drop_first=True)
```

```python
dt=dt.drop(['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime'],axis=1)
```

```python
dt=pd.concat([dt,dt_dummies],axis=1)
```

```python
dt.head()
```

```python
dt.shape
```

```python
x=dt.drop('Attrition',axis=1)
y=dt['Attrition']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

# Object-1

```python
from sklearn import tree
dtree1=tree.DecisionTreeClassifier()
dtree1.fit(x_train,y_train)
y_pred1=dtree1.predict(x_test)
```

# Object-2

```python
dtree2=tree.DecisionTreeClassifier(criterion='entropy',max_depth=4)
dtree2.fit(x_train,y_train)
y_pred2=dtree2.predict(x_test)
```

# Object-3

```python
dtree3=tree.DecisionTreeClassifier(criterion='entropy',max_depth=2,min_samples_leaf=2)
dtree3.fit(x_train,y_train)
y_pred3=dtree3.predict(x_test)
```

```python
from sklearn import metrics
print('Accuracy M1:',metrics.accuracy_score(y_test,y_pred1)) # no estimated parameters
print('Accuracy M2:',metrics.accuracy_score(y_test,y_pred2))
print('Accuracy M3:',metrics.accuracy_score(y_test,y_pred3))
```

# Grid search

```python
from sklearn.model_selection import GridSearchCV
param_grid={"max_depth":[2,3,4,5],
            "max_features":[1,3,5,10],
            "min_samples_split":[2,3,5,10],
            "criterion":["gini","entropy"],
            "min_samples_leaf":[1,2,3,4,5]}
grid_search=GridSearchCV(dtree1,param_grid,cv=5)
grid_search.fit(x_train,y_train)
grid_search.best_params_
```

```python
dtree4=tree.DecisionTreeClassifier(criterion='entropy',max_depth=5,max_features=5,min_samples_leaf=4,min_samples_split=5)
dtree4.fit(x_train,y_train)
y_pred4=dtree4.predict(x_test)
```

```python
print('Accuracy M1:',metrics.accuracy_score(y_test,y_pred1)) # no estimated parameters
print('Accuracy M2:',metrics.accuracy_score(y_test,y_pred2))
print('Accuracy M3:',metrics.accuracy_score(y_test,y_pred3))
print('Accuracy M4:',metrics.accuracy_score(y_test,y_pred4))
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix:",confusion_matrix(y_test,y_pred4))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred4))
```
